```python
cc['ELEVATION'] *= 0.0001  # 높이 차이를 좀 줄이장
dd = np.array(cc.loc[:][['LATITUDE', 'LONGITUDE', 'ELEVATION']])
dd.shape, dd
```

```python
nbrs = neighbors.NearestNeighbors(n_neighbors=7, radius = 1.377, algorithm='brute').fit(dd)
distance, indice = nbrs.radius_neighbors(dd)
```

```python
for i in range(len(distance)):
    if len(distance[i]) < 6:
        continue
    else:
        while len(distance[i]) > 6:
#             print(distance[i])
            kk = np.where(distance[i] == max(distance[i]))
            distance[i] = np.delete(distance[i], kk)
            indice[i] = np.delete(indice[i], kk)
#     a = np.where(distance[i] > 3.)
#     print(a)  # len(distance[i] - len(a)) 7 - 5 = 2
#     print(len(distances[i]) - len(a[0]))
#     distances[i] = distances[i][~np.isin(np.arange(distances[i].size), a)]
```

```python
ind1 = []
ind2 = []
for i in range(len(indice)):
    new_ind1 = [i]*(len(indice[i])-1)
    new_ind2 = []
    for j in range(len(indice[i])):
        if indice[i][j] != i:
            new_ind2 += [indice[i][j]]
    ind1 += new_ind1
    ind2 += new_ind2
print(len(ind1), len(ind2))
```

```python
for j in range(1, 5):
    
    for i in np.where(hh.iloc[:,j] == -9999)[0]:
        
        if (hh.iloc[i, 5] == 6) or (hh.iloc[i, 5] == 25) or (hh.iloc[i-1, j] == -9999) or (hh.iloc[i, j] != -9999):
            continue
        
        if hh.iloc[i+1, j] != -9999:
            hh.iloc[i, j] = round((hh.iloc[i-1, j] + hh.iloc[i+1, j])/2, 4)
            
        elif hh.iloc[i+2, j] != -9999:
            hh.iloc[i, j] = round((hh.iloc[i-1, j]*2 + hh.iloc[i+2, j])/3, 4)
            hh.iloc[i+1, j] = round((hh.iloc[i-1, j] + hh.iloc[i+2, j]*2)/3, 4)
            
        elif hh.iloc[i+3, j] != -9999:
            hh.iloc[i, j] = round((hh.iloc[i-1, j]*3 + hh.iloc[i+3, j])/4, 4)
            hh.iloc[i+1, j] = round((hh.iloc[i-1, j]*2 + hh.iloc[i+3, j]*2)/4, 4)
            hh.iloc[i+2, j] = round((hh.iloc[i-1, j] + hh.iloc[i+3, j]*3)/4, 4)
            
        elif hh.iloc[i+4, j] != -9999:
            hh.iloc[i, j] = round((hh.iloc[i-1, j]*4 + hh.iloc[i+4, j])/5, 4)
            hh.iloc[i+1, j] = round((hh.iloc[i-1, j]*3 + hh.iloc[i+4, j]*2)/5, 4)
            hh.iloc[i+2, j] = round((hh.iloc[i-1, j]*2 + hh.iloc[i+4, j]*3)/5, 4)
            hh.iloc[i+3, j] = round((hh.iloc[i-1, j] + hh.iloc[i+4, j]*4)/5, 4)
```
